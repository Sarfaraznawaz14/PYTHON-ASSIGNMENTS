# 10. How can you tell the difference between break and continue?

```python
# break
i = 1
while i<10:
    i = i+1
    print(i)
    if i == 5:
        break
```

```python
# continue
i = 1
while i<10:
    i = i+1
    print(i)
    if i == 5:
        print('the next numbers printed are because of continue')
        continue
        
```
